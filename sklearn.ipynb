```python
! pip install -U scikit-learn
! pip install numpy
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
home = np.array([100,200,300,400,500,600,700,800])
price = np.array([1000,2000,3000,4000,5000,6000,7000,8000])

x_train,x_test,y_train,y_test=train_test_split(home,price,test_size=0.5,random_state=42,shuffle=False)
# x_train=x_train.reshape(1, -1)
# y_train=y_train.reshape(1, -1)
model = LinearRegression()
model.fit(x_train.reshape(-1, 1),y_train)
res=model.predict(x_test.reshape(-1, 1))
print(x_test)
print(res)
x_test=x_test.reshape(-1, 1)
y_test=y_test.reshape(-1, 1)
print(model.score(x_test,y_test))
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

num_voters = np.array([100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]).reshape(-1, 1)
percentage_votes = np.array([30, 35, 40, 45, 50, 55, 60, 65, 70, 75])

X_train, X_test, y_train, y_test = train_test_split(num_voters, percentage_votes, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(X_test)
print(y_pred)

X_test=X_test.reshape(-1, 1)
y_test=y_test.reshape(-1, 1)
print(model.score(X_test,y_test))

```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

num_voters = np.array([100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]).reshape(-1, 1)
percentage_votes = np.array([30, 35, 40, 45, 50, 55, 60, 65, 70, 75])

X_train, X_test, y_train, y_test = train_test_split(num_voters, percentage_votes, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
plt.title('Linear Regression: Number of Voters vs Percentage of Votes')
plt.xlabel('Number of Voters')
plt.ylabel('Percentage of Votes')
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

temperature = np.array([20, 25, 30, 35, 40, 45, 50, 55, 60, 65]).reshape(-1, 1)
rainfall = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50, 55])

X_train, X_test, y_train, y_test = train_test_split(temperature, rainfall, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
plt.title('Linear Regression: Temperature vs Rainfall')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Rainfall (mm)')
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

age = np.array([25, 30, 35, 40, 45, 50, 55, 60, 65, 70]).reshape(-1, 1)  # Reshape for single feature
insurance_coverage = np.array([100000, 150000, 200000, 250000, 300000, 350000, 400000, 450000, 500000, 550000])

X_train, X_test, y_train, y_test = train_test_split(age, insurance_coverage, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
plt.title('Linear Regression: Age vs Life Insurance Coverage')
plt.xlabel('Age')
plt.ylabel('Life Insurance Coverage ($)')
plt.show()
```
